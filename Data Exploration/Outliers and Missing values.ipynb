```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("D:/4th year/sem8/Data Mining and Warehousing/hotel_bookings.csv")  

missing_cols = df.columns[df.isnull().any()]

plt.figure(figsize=(10, 6))
sns.heatmap(df[missing_cols].isnull(), cbar=True, cmap='magma', yticklabels=False)
plt.title('Heatmap of Missing Values')
plt.xlabel('Columns with Missing Values')
plt.ylabel('Data Entries')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:/4th year/sem8/Data Mining and Warehousing/hotel_bookings.csv")  

exclude_columns = [
    'is_canceled', 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month',
    'arrival_date_week_number', 'is_repeated_guest','adr','agent','company'
]

numerical_df = df.drop(columns=exclude_columns, errors='ignore')  
numerical_df = numerical_df.select_dtypes(include='number')  


plt.figure(figsize=(14, 6))
plt.boxplot([numerical_df[col].dropna() for col in numerical_df.columns], vert=True, patch_artist=True)
plt.xticks(range(1, len(numerical_df.columns) + 1), numerical_df.columns, rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
